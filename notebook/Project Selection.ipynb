## Projects from 'Practical Regression Test Selection with Dynamic File Dependencies'

```python
pn = list(builds['gh_project_name'].unique())
l = [
    'cucumber',
    'joda',
    'retrofit',
    'validator',
    'bval',
    'jxpath',
    'hopper',
    'river',
    'functor',
    'empire',
    'jfreechart',
    'commons',
    'apache',
    'pdfbox',
    'gscollections',
    'closure',
    'log4j',
    'jgit',
    'ivy',
    'jenkins',
    'ant',
    'continuum',
    'guava',
    'camel',
    'jetty',
    'hadoop',
    'zookeeper',
]

matches = []
for candidate in l:
    matches.append((candidate, [x for x in pn if candidate in x]))
matches
```

## Retrofit 

```python
def test(project):
    retrofit = builds[builds['gh_project_name'] == project]
    return len(retrofit[retrofit['test_failures']])

for (group, items) in matches:
    print(group)
    for i in items:
        print(i, test(i))
    print()
    
```
