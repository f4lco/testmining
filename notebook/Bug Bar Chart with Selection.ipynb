```python
import altair as alt
import pandas as pd

alt.renderers.enable('notebook')
```

```python
df = pd.DataFrame([
  ('A', 'X', 0.25),
  ('A', 'Y', 0.50),
  ('B', 'X', 0.75),
  ('B', 'Y', 0.42),
  ('C', 'X', 0.70),
  ('C', 'Y', 0.23),
], columns=['project', 'stage', 'performance'])

df
```

```python
projects = pd.DataFrame(df['project'].unique(), columns=['project'])
projects
```

```python
selection = alt.selection_multi(fields=['project'])
color = alt.condition(selection, 'project:N', alt.value('lightgray'))
projects_selector = alt.Chart(projects).mark_rect().encode(y='project', color=color).add_selection(selection)

projects_bar = alt.Chart(df).mark_bar().encode(
  x='project',
  y='performance',
  color='project',
  column='stage'
).transform_filter(selection)

#projects_selector | projects
print((projects_selector | projects_bar).to_json(validate=False))
```
